```python
#Task 1
import pandas as pd
df = pd.read_csv("titanic.csv")
print(df)
```

```python
#Task 2
df.describe()
```

```python
df.groupby('Survived').agg({'Age':'mean'})
```

```python
df.groupby ('Sex').agg({'Age':'mean','Survived'})
```

```python
#Task 3
# An average of people less than 30 years old survived
import matplotlib.pyplot as plt
df.groupby('Survived')['Age'].mean().plot(kind='bar')
plt.xlabel('Survived')
plt.ylabel('Average Age')
plt.title('Average Age by Survived')
plt.show()
```

```python
# A higher number of people was in the third class than 1st and 2nd
df['Pclass'].plot(kind='hist', bins=10)
plt.xlabel('Sex')
plt.title('Pclass Distribution')
plt.show()
```

```python
#The higher the fare the higher the passenger class with Pclass 1 being the best
df.plot(kind='scatter', x='Pclass', y='Fare')
plt.title('Pclass vs Fare')
plt.show()
```

```python
#People with a lower passenger class of 3 paid less compared to people with a higher passenger class of 1
plt.plot(df['Fare'], df['Pclass'])
plt.xlabel('Fare')
plt.ylabel('Pclass')
plt.title('Fare v Pclass')
plt.show()
```
